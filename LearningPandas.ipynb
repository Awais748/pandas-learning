## Series

A **Pandas Series** is a **1-dimensional labeled array** that can hold any data type.  

Think of it like a **single column in a spreadsheet**.

### Key Points:
- 1-dimensional data structure
- Stores any data type (`int`, `float`, `string`, etc.)
- Has labels called **index**
- Part of the Pandas library



```python
import pandas as pd

data = [100,102,104,200,300,122,213,145,167,189]

series = pd.Series(data, index=['a','b','c','d','e','f','g','h','i','j'])
# series.loc['b'] = 200012 
# print(series[series >= 100])
```

```python
calories = {"Day 1": 1750, "Day 2": 1800, "Day 3": 1900, "Day 4": 2000, "Day 5": 2100}
series = pd.Series(calories)
print(series.loc["Day 3"])
```

## DataFrame

A **Pandas DataFrame** is a **2-dimensional tabular data structure** with **rows and columns**.

Think of it like an **Excel spreadsheet** or a **database table**.

### Key Points:
- 2-dimensional data structure
- Organizes data into rows and columns
- Can store different data types in each column
- Part of the Pandas library

```python
data = {"Name": ["john", "sin", "ash","drake"],
        "Age": [25, 30, 35, 40],
        "City": ["New York", "Los Angeles", "Chicago", "Houston"]}
df = pd.DataFrame(data, index=['Employee 1', 'Employee 2', 'Employee 3', 'Employee 4'])
df["Job"] = ["Developer", "Designer", "Manager", "Analyst"]
new_rows = pd.DataFrame([{"Name": "Emily", "Age": 28, "City": "San Francisco", "Job": "Data Scientist"},
                         {"Name": "Michael", "Age": 32, "City": "Seattle", "Job": "Product Owner"},
                         {"Name": "Sarah", "Age": 29, "City": "Austin", "Job": "UX Designer"},
                         {"Name": "David", "Age": 45, "City": "Miami", "Job": "Project Manager"}
                        ], index=["Employee 5", "Employee 6", "Employee 7", "Employee 8"])
df = pd.concat([df, new_rows])
# print(df.loc['Employee 2'])

print(df)
```

## Data Cleaning

Data cleaning is the process of fixing or removing:
- incomplete data
- incorrect data
- irrelevant data

Around **75% of work done with Pandas** involves data cleaning.

### Common Data Cleaning Tasks
- Removing missing values
- Fixing incorrect data types
- Removing duplicates
- Renaming columns
- Handling empty rows


```python
import pandas as pd

# Load dataset
df = pd.read_csv("pokemon.csv")

# -----------------------------------
# 1. Drop irrelevant columns
# -----------------------------------
df = df.drop(columns=["#"])

# -----------------------------------
# 2. Handle missing data
# -----------------------------------

# Remove rows with missing values
# df = df.dropna()

# OR fill missing values
df = df.fillna("Unknown")

# -----------------------------------
# 3. Fix inconsistent data
# -----------------------------------

# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# -----------------------------------
# 4. Standardize text
# -----------------------------------

# Convert Pokémon names to lowercase
df["Name"] = df["Name"].str.lower()

# -----------------------------------
# 5. Fix data types
# -----------------------------------

# Convert Total column to integer
df["Total"] = df["Total"].astype(int)

# -----------------------------------
# 6. Remove duplicate values
# -----------------------------------

df = df.drop_duplicates()

```
